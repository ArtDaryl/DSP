# PART 1

Solve for the magnitude and angle of the linear transformations given. Provide necessary solutions numerically (hand-solved) and computationally (using a Python program). Also provide the necessary solutions vector plots of the given matrices and their transformations (nos. 1-5.)

\begin{align} H = \frac{1}{\sqrt{2}}\begin{bmatrix} 1,1\\1,-1 \end{bmatrix}; Y = \begin{bmatrix} 0,-i\\i,-0 \end{bmatrix}; O = \begin{bmatrix} 1,0\\1,1 \end{bmatrix}\end{align}

1. H ⋅ Y
2. Y ⋅ H
3. H ⋅ H
4. Y ⋅ H ⋅ O
5. H ⋅ Y ⋅ H ⋅ O

```python
H = (1 / np.sqrt(2)) * np.array([[1, 1],[1,-1]])

i = 1
Y = np.array([
    [0,-i],
    [i,-0]])

O = np.array([
    [1,0],
    [1,1]])
```

```python
HY =  np.dot(H,Y)
magnitude = np.linalg.norm(HY)
rad = np.arctan2(HY[1, 0], HY[0, 0])
angle = np.degrees(rad)
print(HY)
print("Magnitude = ",magnitude,"  Angle =", angle)

fig, ax = plt.subplots()
ax.quiver(0, 0, HY[0,1], HY[1,0], angles='xy', scale_units='xy', scale=1, label='HY',color='blue')
ax.quiver(0, 0, H[0,1], H[1,0], angles='xy', scale_units='xy', scale=1, label='H',color='red')
ax.quiver(0, 0, Y[0,1], Y[1,0], angles='xy', scale_units='xy', scale=1,label='Y',color='green')
ax.quiver(0, 0, magnitude, 0, angles='xy', scale_units='xy', scale=1, label='Magnitude', color='black')
ax.set_xlim([-1.5,1.5])
ax.set_ylim([-1.5,1.5])
plt.grid()
plt.legend()
plt.show()
print("Plot of HY")
```

# PART 3

Determine whether the resulting linear transformations are linearly dependent. Provide necessary solutions both numerically using determinants (hand-solved) and computationally (using a Python program).

\begin{align}
1.)\begin{pmatrix} 5&0&0\\0&5&0\\0&0&5 \end{pmatrix}    \begin{pmatrix} 1&0&5\\2&7&6\\6&4&7 \end{pmatrix}
2.) \begin{pmatrix} 1&2&6\\3&15&4\\2&10&3 \end {pmatrix}   \begin{pmatrix} 5&2&4\\6&2&4\\0&1&1 \end{pmatrix}
  \end{align}

```python
A = np.array ([
    [5,0,0],
    [0,5,0],
    [0,0,5],
])
B= np.array ([
    [1,0,5],
    [2,7,6],
    [6,4,7],
])

dot = np.dot(A, B)

if np.all(dot == 0):
    print("Linearly dependent")
else:
    print("Linearly independent")
```

```python
C = np.array ([
    [1,2,6],
    [3,15,4],
    [2,10,3],
])
D = np.array ([
    [5,2,4],
    [6,2,4],
    [0,1,1],
])

dot = np.dot(C, D)

if np.all(dot == 0):
    print("Linearly dependent")
else:
    print("Linearly independent")
```

Plot the signal corresponding to the following vectors whereas 𝑇 is the time vector and 𝐺 is the vector corresponding to the amplitudes of the signal.

\begin{align}
 \large T = \begin{pmatrix}0 & \frac{π}{4} & \frac{π}{2} & \frac{3π}{4} \end{pmatrix};
 G =  \begin{pmatrix} 5 & 3 & 0 & -3 & 5 \end{pmatrix}

```python
T = np.array([0, np.pi/4, np.pi/2, 3*np.pi/4])
G = np.array([5, 3, 0, -3])

plot = G
plt.stem(T, plot, use_line_collection=True)
plt.xlabel('Time (T)')
plt.ylabel('Amplitude')
plt.title('Signal Plot')
plt.grid(True)
plt.show()
```
