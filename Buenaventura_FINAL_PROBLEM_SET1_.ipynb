Part 1. [200 pts] Compute for the Energy and Power of the following signals in terms of per unit use the proper formula whether the signal is continuous or discrete.

\begin{align*}
\begin{array}{cc}
\text{a. } \sin[n] & \text{c. } 2[\cos(t) + i\sin(t)] \\
\text{b. } \cos(t) & \text{d. } y[n] = \frac{2}{\sqrt{\pi}}-e^{in} \\
\end{array}
\end{align*}

```python
print("sin[n]")
N_values = np.linspace(-1000, 1001, 10)
x_n = np.sin(N_values)
E = np.sum(np.abs(x_n)**2)
P = E / len(N_values)

print(f"Energy = {E}")
print(f"Power = {P}")
print("Discrete")
```

```python
T = np.pi
T_values = np.linspace(-T,T,10)
print("cos(t)")
E1 = (1/2) * T_values  + (1/4) * np.sin(2*T_values)
P1 = (1/2*T)*E1
print(f"Energy = {E1}")
print(f"Power = {P1}")
print("Continuous")
```

```python
print("2[cos(t)+isin(t)]")
E2 = 3/2 * T_values + 5/2 * np.sin(2*T_values) - 3/2 * np.cos(2*T_values)
P2 = (1/2*T)*E2
print(f"Energy = {E2}")
print(f"Power = {P2}")
print("Continuous")
```

```python
print("y[n]=(2/√π)−e^in")
y_n = 2 / np.sqrt(np.pi) - np.exp(1j * N_values)
E3 = np.sum(np.abs(y_n)**2)
P3 = np.abs(y_n[0])**2
print(f"Energy = {E3}")
print(f"Power = {P3}")
print("Discrete")
```
