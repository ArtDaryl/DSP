#Part 1
Compute and plot (you may or may not use software) the following complex numbers. You should have a total of three (3) graphs. First graph includes numbers 1 to 4. Second graph includes numbers 5 to 9. The third graph includes only number 10.

\begin{align}
z_1 &= 1 - 5i, &\quad z_6 &= z_2 \cdot (z_1 - z_3) \\
z_2 &= 3 + 2i, &\quad z_7 &= \frac{z_4}{z_1} \\
z_3 &= -7 + i, &\quad z_8 &= \frac{z_4}{z_2} \cdot (z_1 + z_4) \\
z_4 &= -1.6i, &\quad  z_9 &= \frac{z_1}{z_3} + \frac{2z_4}{3} \\
z_5 &= z_1 \cdot z_1, &\quad \omega &= e^{z_4} \\
\end{align}

```python
num14 = [1 - 5j, 3 + 2j, -7 + 1j, -1.6j]
real = [num.real for num in num14]
imaginary = [num.imag for num in num14]

plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Graph of 1-4')
plt.plot(real, imaginary, c='b', markerfacecolor='g', marker='o', linestyle='-',label='Complex Numbers')
plt.legend()
plt.xlim(-10, 9)
plt.ylim(-10, 9)
plt.grid(True)
plt.show()
```

```python
num59 = [-4 - 10j, 36 - 6j, 0.308 - 0.062j, -2.314 + 1.009j, -0.24 - 0.387j]
real = [num.real for num in num59]
imaginary = [num.imag for num in num59]

plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Graph of 5-9')
plt.plot(real, imaginary, c='b', markerfacecolor='g', marker='o', linestyle='-',label='Complex Numbers')
plt.legend()
plt.xlim(-10, 50)
plt.ylim(-50, 50)
plt.grid(True)
plt.show()
```
